```python
import pandas as pd

def has_identical_group(df):
    def check_identical(group):
        return group.nunique() == 1

    # Group by 'source' and 'past_split_id' and check if all values in 'Prototype_id' are identical
    grouped = df.groupby(['source', 'past_split_id'])['Prototype_id'].apply(check_identical)

    # Return True if there is at least one True in grouped
    return grouped.any()

# Suppose df is your DataFrame
print(has_identical_group(df))
print(has_identical_group(filtered_df))
```

```python
def transform_prototype(group):
    protos = group["Prototype_num"].values
    new_protos = []
    last_proto = None
    for proto in protos:
        if last_proto is None or last_proto != proto:
            new_protos.append("TRUE")
        else:
            new_protos.append("FALSE")
        last_proto = proto
    group["Segment_bool"] = new_protos
    return group

# apply the transform_prototype function to each group of rows with the same "episode_id" and "Prototype_id"
filtered_df_bool = filtered_df.groupby(['source', "Prototype_id"]).apply(transform_prototype)
```

```python

# Load the DataFrame containing the start and end times
#df = pd.read_csv("/content/drive/MyDrive/TAL_FOCUS/Episodes/episodes_TAL/TAL_TAGGED_020040066080099140160.csv",index_col = [0])
#df = df[df["tier_name"] == "words"]
#df = df[df['source'].str.contains(r"^SBC\d+$", regex=True)]
#df['source'].unique()


words_files = []
protoype_label_files = []
wav_files = []



# Get a list of all the wave files in the directory and its subdirectories
#file_list = glob.glob('/content/drive/MyDrive/TAL_FOCUS/Episodes/episodes_TAL/*.wav')

file_list = file_list


# specify your path
path_to_save = "/home/eranbe/dharelg/datasets/whisper_dataset/prototype_dataset/TAL_4prototype_tmpfiles_4creating_dataset/garbage5" 
# check if the directory exists
if not os.path.isdir(path_to_save):
    # if the directory doesn't exist, create it
    os.makedirs(path_to_save)



# Iterate over the list of wave files
for i,file_wave in enumerate(file_list):
    
    print(file_wave)
    
    

    
    # Open the MP3 file using PyDub
    audio = AudioSegment.from_file(file_wave, format="mp3") 
    file_mp = file_wave
    file_wave_tmp = path_to_save + file_mp.split("/")[-1].split(".")[0]
    file_wave = file_wave_tmp +f"_{i}.wav"

    audio.export(file_wave, format="wav")
    

    
    
    
    with wave.open(file_wave, 'rb') as f:
        # Get the wave file properties
        nchannels = f.getnchannels()
        sampwidth = f.getsampwidth()
        framerate = f.getframerate()
        nframes = f.getnframes()
        # Iterate over the groups in the DataFrame
        
    
    
    turn_df_name = file_mp.split("/")[-1].split(".")[-2]
    
    
    turn_df = df[df['source'] == float(turn_df_name)]

    
    for turn_id, group in turn_df.groupby('past_split_id'):
        # Get the start and end times for this group
        
        
        start_time = group['xmin'].iloc[0]
        end_time = group['xmax'].iloc[-1]
        
        # Calculate the start and end frame indices
        start_frame = int(start_time * framerate)
        end_frame = int(end_time * framerate)

        

        
        # Crop the wave file
        
     
        wav_file = path_to_save+'/wav/'+str(int(turn_id))+'.wav'
        wav_folder = path_to_save+'/wav/'
        
        
        # check if the directory exists
        if not os.path.isdir(wav_folder):
            # if the directory doesn't exist, create it
            os.makedirs(wav_folder)

        
        
        
        
        
        with wave.open(wav_file, 'wb') as f_out:
            f_out.setparams((nchannels, sampwidth, framerate, end_frame - start_frame, 'NONE', 'not compressed'))
            with contextlib.closing(wave.open(file_wave, 'rb')) as f_in:
                f_in.setpos(start_frame)
                f_out.writeframes(f_in.readframes(end_frame - start_frame))
        
        
        
        
        
        
        
        boolean_folder = path_to_save+'/boolean/'
        protoype_label_file = path_to_save+'/boolean/'+str(turn_id)+'.txt'
        # check if the directory exists
        if not os.path.isdir(boolean_folder):
            # if the directory doesn't exist, create it
            os.makedirs(boolean_folder)

        
       
        with open(protoype_label_file, 'w') as f:
          f.write((group['label'].apply(lambda x: ''.join(str(x))).to_string()))

        
        
        
        
        
                
        
        word_folder = path_to_save+'/words/'
        word_file = path_to_save+"/words/"+str(turn_id)+".txt"
        
        # check if the directory exists
        if not os.path.isdir(word_folder):
            # if the directory doesn't exist, create it
            os.makedirs(word_folder)
            
            
        
        with open(word_file, 'w') as f:
          f.write((group['words'].apply(lambda x: ''.join(str(x))).to_string())) 
        
        
        words_files.append(word_file)
        protoype_label_files.append(protoype_label_file)
        wav_files.append(wav_file)
```
